```python
# create a new df to store the filtered slices:

df = pd.DataFrame(columns=["case", "slice", "distance"])
```

```python
for case, slice, distance in zip(output['case'], output['slice'], output['distance']):
    for case_id, start, end in zip(ranges_df['case_id'], ranges_df['start_slice'], ranges_df['end_slice']):
        if case == case_id and start <= slice <= end:
            df.loc[len(df)] = {
                "case": case,
                "slice": slice,
                "distance": distance
            }
```
